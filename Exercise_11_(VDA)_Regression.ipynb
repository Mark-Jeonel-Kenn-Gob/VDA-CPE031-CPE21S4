#Sample Problem
Given the following sets of data:

x = [0.5, 1.00, 1.30, 1.60, 2.00, 2.20, 2.50, 3.10, 3.90, 4.40]

y = [1.90, 1.50, 1.20, 1.00, 0.80, 0.78, 0.65, 0.46, 0.30, 0.23]

Use linear model and interpret your data.

```python
# Sample Problem
# Use linear model and interpret your data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.5, 1, 1.3, 1.6, 2, 2.2, 2.5, 3.1, 3.9, 4.4])
y = np.array([1.9, 1.5, 1.2, 1, 0.8, 0.78, 0.65, 0.46, 0.3, 0.23])

# Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

# Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

# Plot the original data points and the regression line
plt.scatter(x, y, color='green', s=30)
plt.plot(x, y_pred, color='purple')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.title('Sample Problem Linear Regression', fontweight='bold', color='orange')
plt.grid(axis = 'y')
plt.legend(['Original Data', 'Regression Line'])
plt.show()

print(f'Slope: {slope}')
print(f'Y-intercept: {intercept}')
```

#***Interpretation of data:***
The regression line equation following (y = A + Bx) is (y =  1.79 -0.40x)

Furthermore, the slope represents the change in y for a one-unit change in x. In this case, for every one-unit increase in x, y decreases by approximately 0.40. Following that, the y-intercept represents the predicted value of y when x is 0. With the given sample data, when x is 0, y is predicted to be approximately 1.79. Moreover, the negative slope indicates a negative linear relationship between x and y,meaning as x increases, y tends to decrease.
