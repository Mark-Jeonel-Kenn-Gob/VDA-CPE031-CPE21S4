```python
import pandas as pd

file_path = '/content/drive/MyDrive/Regression_Sheet.csv'

df = pd.read_csv('/content/drive/MyDrive/Regression_Sheet.csv')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

# Calculate the predicted values of y using th eregression line equation
y_pred = slope * x + intercept

# Plot the original data points and the regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = df['x (Proficiency Exam Score)'].head(9)
y = df['y (Course Grade)'].head(9)

# Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

# Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

# Plot the original data points and the regression line
plt.scatter(x, y, color='orange')
plt.plot(x, y_pred, color='magenta')
plt.xlabel('x (Proficiency Exam Score)')
plt.ylabel('y (Course Grade)')
plt.title('Regression Line', color='Teal')
plt.legend(['Original Data Points', 'Regression Line'])
plt.show()

print(f'Slope: {slope}')
print(f'Y-intercept: {intercept}')
```
