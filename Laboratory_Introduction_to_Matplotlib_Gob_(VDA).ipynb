#Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show
```

#Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show
```

#Histogram

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data)
plt.title("Histogram Plot Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show
```

#Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show
```

#Box Plot

```python
import matplotlib.pyplot as plt

data = [[1.5]*10 + [2]*10 +[3]*10]
plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show
```

#Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show
```

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/mdc.csv'
df = pd.read_csv(file_path, encoding='latin-1')

print(df.head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas as pd

entity_counts = df['entity'].value_counts()

colors = ['red', 'blue'] #Specifies color for each slice of pie chart

plt.figure(figsize=(6, 5)) #Specifies the figure size (6x5)
plt.pie(entity_counts, labels=entity_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
#Includes auto percent as seen inside each slice
plt.title("Distribution of Entities (Marvel vs DC)", color='green', fontweight = 'bold')
#Title with color specified, as well as in boldface
plt.legend(title = "Marvel vs DC Films")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive

plt.figure(figsize=(8, 6)) #Specifies figure size
plt.hist(df['year'], bins=45, color='magenta') #Specifies number of bins and color
plt.title('Histogram of Film Release Dates', style='italic', color='green')  #Title italicized colored green
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

director_counts = df['director'].value_counts()

plt.figure(figsize=(10, 6))  #Specifies figure size
director_counts.plot(kind='bar', color='orange') #Specifies bar color as orange
plt.title('Number of Films per Director', color='green', fontweight='bold')

plt.xlabel('Director', color='blue', style='italic')
plt.ylabel('Number of Films', color='red', style='italic')
plt.xticks(rotation=90, ha='right', fontsize=8)  #Rotate x-axis labels to 90deg for better readability
plt.grid(axis='y', linestyle=':', linewidth=0.5) #Show horizontal gridlines with dotted style
plt.show()
```
