```python
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/heart_2022_with_nans.csv'
df = pd.read_csv(file_path, encoding='latin-1')
df
```

#Questions:
1. What is the distribution of the general health of the survey participants?
2. What is the correlation between the participants' sleeping hours and BMI?


```python
#1. What is the distribution of the general health of the survey participants?

plt.figure(figsize=(9, 7))
colors = ['hotpink', 'orange', 'yellow', 'teal', 'skyblue']
gh = df['GeneralHealth'].value_counts()
plt.pie(gh, labels=gh.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.legend(title = "General Health", loc='lower right')
plt.title("Distribution of Participants' General Health", fontweight='bold')
plt.axis('equal')
plt.show()

# OBSERVATION:
# Answering the question, assessing the Pie Chart produced lets me deduce the following spread of values for the distribution of the general health of the survey participants:
# Poor = 4.4%; Fair = 13.6%; Good = 32.3%; Very Good = 33.4%; Excellent = 16.2%
# Based on this, I can conclude that majority of the participants (which are US residents based on the given analysis and information under the States variable) can be classified as
# Good or Very Good, with the minority being those with Poor general health, followed by those with a Fair rating. Those with Excellent general health are the median or mean class.
```

```python
#2. What is the correlation between the participants' sleeping hours and BMI?

plt.figure(figsize=(10, 5))
SleepToBMI = df.groupby('SleepHours')['WeightInKilograms'].sum()

plt.bar(SleepToBMI.index, SleepToBMI.values, color='olive')
plt.xlabel('Daily Sleeping Hours')
plt.ylabel('Weight in kg')
plt.title('Weight to Daily Sleeping Hours', fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# OBSERVATION:
# Answering the question, I created a bar chart highlighting the relationship between the variables Sleep Hours and Weight in Kilograms. This helped me determine that the average sleeping hours
# of 8 hours is directly related to higher and normal BMIs. This also suggests that those that acquire more or less of the average 8 hours have lower BMIs.
# The bar chart/graph produced appears with a distribution close to a normal distribution (bell curve), implying that there is positive correlation between the two variables tested.
```
