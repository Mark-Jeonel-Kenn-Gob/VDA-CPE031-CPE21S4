Matplot Labels and Title - You can use the xlabel() and ylabe()

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)

plt.title("Distance Travelled Everyday", color="red")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.show()
```

Matplot Grid - You can use grid() to add grid lines

```python
#Display gridlines
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)

plt.title("Distance Travelled Everyday", color="red")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid()
plt.show()
```

```python
#Select gridlines to display
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)

plt.title("Distance Travelled Everyday", color="red")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid(axis='y')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)

plt.title("Distance Travelled Everyday", color="red")
plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.grid(axis='x', linestyle='--', linewidth=0.5)
plt.show()
```

Show Multiple Plots

```python
#plot 1
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1) # NumOfRows | NumOfColums | Position/Index
plt.plot(xpoints, ypoints)

#plot 2
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(xpoints, ypoints)

plt.grid(axis='x', linestyle='--', linewidth=0.5)
plt.show()
```

Matplot Scatter - The scatter function plots one dot for each observation

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.scatter(xpoints, ypoints)

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([10, 20, 30, 40])

plt.scatter(xpoints, ypoints, color='magenta')

plt.show()
```

Matplotlib Bars - You can use the bar() to draw bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Student 1", "Student 2", "Student 3"])
y = np.array([90, 85, 86])

colors = ['red', 'magenta', 'green']

plt.bar(x, y, color=colors)
plt.show()
```

Matplotlib Pie Chart - You can use pie() to draw pie chart

```python
import matplotlib.pyplot as plt
import numpy as np

g = np.array([35, 25, 25, 15])
grader = ["Grade 1=35%", "Grade 2=25%", "Grade 3=25%", "Grade 4=15%"]

plt.pie(g, labels=grader)
plt.legend(title = "Grader")
plt.show()
```
