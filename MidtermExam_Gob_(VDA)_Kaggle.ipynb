##Indicators of Heart Disease 2022 (With nans)

Questions:
1. What is the distribution of the general health of the survey participants?
2. What is the correlation between the participants' sleeping hours and BMI?


```python
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/heart_2022_with_nans.csv'
df = pd.read_csv(file_path, encoding='latin-1')
df
```

```python
#1. What is the distribution of the general health of the survey participants?

plt.figure(figsize=(9, 7))
colors = ['hotpink', 'orange', 'yellow', 'teal', 'skyblue']
gh = df['GeneralHealth'].value_counts()
plt.pie(gh, labels=gh.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.legend(title = "General Health", loc='lower right')
plt.title("Distribution of Participants' General Health", fontweight='bold')
plt.axis('equal')
plt.show()

# OBSERVATION:
# Answering the question, assessing the Pie Chart produced lets me deduce the following spread of values for the distribution of the general health of the survey participants:
# Poor = 4.4%; Fair = 13.6%; Good = 32.3%; Very Good = 33.4%; Excellent = 16.2%
# Based on this, I can conclude that majority of the participants (which are US residents based on the given analysis and information under the States variable) can be classified as
# Good or Very Good, with the minority being those with Poor general health, followed by those with a Fair rating. Those with Excellent general health are the median or mean class.
```

```python
#2. What is the correlation between the participants' sleeping hours and BMI?

plt.figure(figsize=(10, 5))
SleepToBMI = df.groupby('SleepHours')['WeightInKilograms'].sum()

plt.bar(SleepToBMI.index, SleepToBMI.values, color='olive')
plt.xlabel('Daily Sleeping Hours')
plt.ylabel('Weight in kg')
plt.title('Weight to Daily Sleeping Hours', fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# OBSERVATION:
# Answering the question, I created a bar chart highlighting the relationship between the variables Sleep Hours and Weight in Kilograms. This helped me determine that the average sleeping hours
# of 8 hours is directly related to higher and normal weight. This also suggests that those that acquire more or less of the average 8 hours have lower weight.
# The bar chart/graph produced appears with a distribution close to a normal distribution (bell curve), implying that there is positive correlation between the two variables tested.
```

##Indicators of Heart Disease 2022 (No nans)
Questions:
1. How does the number of physical health days relate to a person's BMI?
2. How does the number of mental health days affect daily sleep hours?


```python
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/heart_2022_no_nans.csv'
df2 = pd.read_csv(file_path, encoding='latin-1')
df2
```

```python
#1. How does the number of physical health days relate to a person's BMI?

plt.figure(figsize=(8, 6))
plt.scatter(df2['BMI'], df2['PhysicalHealthDays'], alpha=0.25, color='maroon', marker='o')
plt.xlabel('BMI')
plt.ylabel('Physical Health Days')
plt.title('Relationship between BMI and Physical Health Days', fontweight='bold')
plt.grid(axis='y')
plt.show()

# OBSERVATION:
# Examining the scatter plot generated, we can get an idea of the relationship between BMI and Physical Health Days.
# The scatter plots is a great starting point for exploring the relationship between the two variables, and it helps
# me deduce that there is a positive correlation between BMI and Physical Health Days (monthly) as, despite the noise
# present in certain parts of the scatter plot, higher frequencies in the monthly physical health days show smaller BMIs,
# while lower frequencies of physical health days correspond to higher BMIs.
```

```python
#2. How does the number of mental health days affect daily sleep hours?

sleep_mental_health = df2.groupby('SleepHours')['MentalHealthDays'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.plot(sleep_mental_health['SleepHours'], sleep_mental_health['MentalHealthDays'], marker='x', linestyle=':', color='magenta', ms=10, mec='g')
plt.xlabel('Daily Sleeping Hours')
plt.ylabel('Average Mental Health Days')
plt.title('Relationship between Sleep Hours and Mental Health Days', fontweight='bold')
plt.grid(axis='y')
unique_sleep_hours = sorted(df2['SleepHours'].unique())
plt.xticks(unique_sleep_hours)
plt.tight_layout()
plt.show()

# OBSERVATION:
# After generating the line graph, I can arrive at the following things regarding the relationship between sleep hours and mental health days.
# Having known that 8 hours is generally regarded as the average amount of sleep for an individual, we could notice a dip in mental health days around the
# 8 hours of sleep marker. This suggests a lower need for mental health days when adequate sleep is acquired. Other notable markers are those with
# high sleep hours and high mental health days, which would suggest responses from individuals with sedentary lifestyles or with serious or severe health
# conditions. These all lead me to conclude that a positive correlation between the two variables exists.

# Take note, however, that the provided dataset would not suffice to create solid conclusions without more complex data analysis methods, such as a full-blown research methodology.
# This visualization, along with the previous ones for the .csv file, only help us create initial visualizations for the probable meaning or interpretation of the dataset and its variables.
```
