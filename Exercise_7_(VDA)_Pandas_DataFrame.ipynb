**3. Creating a dataframe from a dictionary of lists:** Manually create a dictionary where each value is composed of a list from your original dataset, then load it into a dataframe, before printing it. You are required to provide at least five (5) observations in your list.

```python
#3. Creating a dataframe from a dictionary of lists.
#Manually create a dictionary where each value is composed of a list from your original dataset, then load it into a dataframe, before printing it.
#You are required to provide at least five (5) observations in your list.

import pandas as pd
data_dict = {
    'Pokedex Number': df['Pokedex Number'].tolist()[:5],
    'Name': df['Name'].tolist()[:5],
    'Type1': df['Type1'].tolist()[:5],
    'Type2': df['Type2'].tolist()[:5],
    'Total': df['Total'].tolist()[:5],
    'HP': df['HP'].tolist()[:5],
    'Attack': df['Attack'].tolist()[:5],
    'Defense': df['Defense'].tolist()[:5],
    'Sp_attack': df['Sp_attack'].tolist()[:5],
    'Sp_defense': df['Sp_defense'].tolist()[:5],
    'Speed': df['Speed'].tolist()[:5],
    'Generation': df['Generation'].tolist()[:5],
    'Legendary': df['Legendary'].tolist()[:5],
}

new_df = pd.DataFrame(data_dict)
new_df
```

**4. Creating a dataframe from a list of dictionaries:** Manually create a list of dictionaries  from your original dataset, then pass it into a dataframe, before printing it. You are required to provide at least five (5) observations in your list.

```python
#4. Creating a dataframe from a list of dictionaries.
#Manually create a list of dictionaries  from your original dataset, then pass it into a dataframe, before printing it.
#You are required to provide at least five (5) observations in your list.

import pandas as pd
data_list = []
for i in range(5):
  data_list.append({
      'Pokedex Number': df['Pokedex Number'][i],
      'Name': df['Name'][i],
      'Type1': df['Type1'][i],
      'Type2': df['Type2'][i],
      'Total': df['Total'][i],
      'HP': df['HP'][i],
      'Attack': df['Attack'][i],
      'Defense': df['Defense'][i],
      'Sp_attack': df['Sp_attack'][i],
      'Sp_defense': df['Sp_defense'][i],
      'Speed': df['Speed'][i],
      'Generation': df['Generation'][i],
      'Legendary': df['Legendary'][i],
  })

new_df_list = pd.DataFrame(data_list)
new_df_list
```

**6. Selecting dataframe rows:** Execute a method that would allow you to select a single and multiple dataframe rows using panda indexing and python indexing.

```python
#6. Selecting dataframe rows.
#Execute a method that would allow you to select a single and multiple dataframe rows using panda indexing and python indexing.

#Selecting a single row using pandas indexing (iloc)
row_at_index_0 = df.iloc[0]
row_at_index_0
```

```python
#Selecting multiple rows using pandas indexing (iloc)
rows_from_index_1_to_3 = df.iloc[1:4]
rows_from_index_1_to_3
```

```python
#Selecting multiple rows using python indexing (list of indices)
rows_at_indices_0_and_2 = df.iloc[[0, 2]]
rows_at_indices_0_and_2
```
